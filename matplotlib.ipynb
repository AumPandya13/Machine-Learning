```python
import matplotlib.pyplot as plt
```

```python
y_views = [534,689,258,401,724,689,350]
f_views = [123,342,700,304,405,650,325]
t_views = [202, 209, 176, 415, 824, 389, 550]
days = range(1,8)
```

```python
plt.plot(days, y_views, label = 'Youtube Views', color = 'r', marker = 'o', markerfacecolor = 'b', linewidth = 2)
plt.plot(days, f_views, label = 'Facebook Views', color = 'b', marker = 'o', markerfacecolor = 'orange', linewidth = 2)
plt.plot(days, t_views, label = 'Twitter Views', color = 'black', marker = 'o', markerfacecolor = 'g', linewidth = 2)
plt.xlabel('Day No. ')
plt.ylabel('views')
plt.legend(loc = 'lower right')
plt.grid(True, linewidth = 1, color='r', linestyle = '--')
plt.title('Daily views for Marketing channels')
plt.show()
```

Scatter Plot

```python
y_views = [534,689,258,401,724,689,534,545,435,423,789,905,561]
days = range( 1, 14)

plt.scatter(days, y_views, label= 'youtube Views')
plt.xlabel('Day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')
plt.title('Daily views for Marketing channels')
plt.show()
```

Bar Plot

```python
y_views = [534,689,258,401,724,689,534,545,435,423,789,905,561]
f_views = [250,350,150,230,400,325,125,500,300,475,234,984,550]
x = range( 1, 14)

plt.bar([a-0.25 for a in x], y_views, width=0.25, label= 'youtube Views', align='edge')
plt.bar([a+0.25 for a in x], f_views, width = -0.25, label= 'facebook Views', align='edge')
plt.xlabel('Day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')
plt.xticks(x) 
plt.title('Daily views for Marketing channels')
plt.tight_layout()
plt.show()
```

Histogram

```python
# Sample data
data = [23, 45, 56, 78, 90, 34, 67, 55, 88, 44, 77, 33, 66, 89, 100, 25, 48]

# Plotting the histogram
plt.figure(figsize=(8, 6))  # Setting figure size
plt.hist(data, bins=10, color='skyblue', edgecolor='black')  # Adjusting number of bins and color
plt.xlabel('Values')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.title('Histogram of Sample Data')  # Title of the plot
plt.grid(True)  # Showing grid
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

Pie Charts

```python
# Sample data
labels = ['Label A', 'Label B', 'Label C', 'Label D', 'Label E']
sizes = [25, 35, 20, 10, 10]  # Percentages adding up to 100 for each label

# Plotting the pie chart
plt.figure(figsize=(8, 6))  # Setting figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange'])
plt.title('Pie Chart Example')  # Title of the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Text annotation
total = sum(sizes)
plt.text(0, 0, f'Total: {total}', fontsize=12, color='black', ha='center')

# Additional details for each section
for i, (label, size) in enumerate(zip(labels, sizes)):
    percentage = sizes[i] / total * 100
    plt.text(1.4, 0.3 - i * 0.2, f'{label}: {size} ({percentage:.1f}%)', fontsize=10, color='black', ha='left')

plt.tight_layout()  # Adjust layout
plt.show()
```

```python
import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add nodes
G.add_nodes_from([1, 2, 3, 4, 5])

# Add edges
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 2)])

# Draw the graph
pos = nx.circular_layout(G)  # Define the positions of the nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500, font_weight='bold', font_size=12, edge_color='gray')

# Show the graph
plt.title('Network Diagram')
plt.axis('off')  # Hide the axis
plt.show()
```

Heat Map

```python
import numpy as np

# Generate sample data (a 5x5 grid)
data = np.random.rand(5, 5)

# Plot heatmap
plt.figure(figsize=(8, 6))
heatmap = plt.imshow(data, cmap='YlGn')  # Choose a colormap ('YlGn' in this case)
plt.colorbar(heatmap)  # Add colorbar to show the scale
plt.title('Heatmap Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(5), ['A', 'B', 'C', 'D', 'E'])  # Customize x-axis labels
plt.yticks(np.arange(5), ['W', 'X', 'Y', 'Z', 'V'])  # Customize y-axis labels
plt.grid(False)  # Hide gridlines

plt.show()
```
