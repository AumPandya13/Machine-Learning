```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
headers_data = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", type(headers_data))
```

```python
file_path = "C:/Users/aumpa/Documents/ML/batwebs/bowling.csv"
df = pd.read_csv(file_path, names=headers_data)
df.head()
```

```python
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
missing_data = df.isnull()
missing_data
```

```python
for column in headers_data:
  print(column)
  print(missing_data[column].value_counts())
  print(" ")
```

```python
# Handling missing values: Imputation or removal
# Example: Impute missing numeric values with mean, and categorical values with mode
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(exclude=np.number).columns.tolist()
```

```python
# Impute missing numeric values with mean
for col in numeric_cols:
    df[col] = df[col].astype(float)
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
# Impute missing categorical values with mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Convert numeric columns to appropriate data types if needed
numeric_cols_except_price = [col for col in numeric_cols if col != 'price']
df[numeric_cols_except_price] = df[numeric_cols_except_price].astype(float)
```
